```python
images_dir = Path('../frontend/public/images')
count = 1
try:
    # Fetch products from the MongoDB collection
    products = collection.find()

    for product in products:
        product_folder = images_dir / str(product["_id"])

        # Create the folder for the product if it doesn't exist
        product_folder.mkdir(parents=True, exist_ok=True)

        # Save each image (if not already saved)
        for index, image_buffer in enumerate(product["images"]):
            image_path = product_folder / f"image{index + 1}.jpg"

            # Check if the image already exists to prevent duplicates
            if not image_path.exists():
                # Convert the image buffer (assuming it's a binary buffer) to an image and save
                image = Image.open(io.BytesIO(image_buffer))
                image.save(image_path)
        print("Count: ", count,  " --- Images of ", product["_id"], " saved \n")
        count = count + 1

    # Respond with the products (or any other desired result)
    print("Products and images saved successfully.")
except Exception as e:
    print(f"Error: {str(e)}")
```
